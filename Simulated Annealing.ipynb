# 模拟退火法


```python
# Q1

from __future__ import division  
import numpy as np  
import matplotlib.pyplot as plt  
import math  

def generatevalue (uu,vv,ww,yy):
    value = ((uu*math.pow(vv,2)+math.sin(math.pi*ww))*math.sin(vv/(0.01 + math.pow(uu,2)))*
             math.sin(math.pi*ww/2) + math.pow(uu,3)*math.pow(vv,2)*ww*
             math.sin(math.pow(vv,3)/(0.001*math.sin(math.pi*ww/2)+math.pow(uu,4)+math.pow(ww-1,2))))*yy
    return value

def Q1():
    T = 100  #initial temperature
    Tmin = 0.01 #turminate temperature
    u = 0  
    ubest = 0
    z = generatevalue(u,1,1,1)
    zbest = generatevalue(u,1,1,1)
    t = 0  # time spend
    t1 = 0 # on which itteration gets the best solution 
    dr = 0.99
    Tgraph = []
    ugraph = []
    zgraph = []
    pgraph = []
     
    while T > Tmin:
        z = generatevalue(u,1,1,1)
        uNew = u + np.random.uniform(low = -0.2, high = 0.2)
        if ((uNew<=1) & (uNew>=-1)):    
            zNew = generatevalue(uNew,1,1,1) #create new z value
            if zNew > z:
                u = uNew #always accept
            else:
                p = math.exp((zNew -z)/T)
                r = np.random.uniform(low = 0, high = 1)
                #print ("p: ", p, "   r: ",r)
                if r < p:
                    u = uNew  #accept based on a ratio
        t += 1
        T= T*dr #cooling
        
        #print ("T: ", T, "  u: ", u, "  z: ", z)
        if zNew > zbest:
            zbest = zNew
            ubest = uNew
            t1 = t
               
        Tgraph.append(T)
        ugraph.append(u)
        zgraph.append(z)

    # draw changes of tempeture, z and u
    print("Question 2a:")    
    print("Itteration: ", t, "  Tempeture:",T, "  u best:", ubest, "  z best: ", zbest, "   get on iteration: ", t1)   
    plt.figure(1)
    plt.plot(np.array(Tgraph))
    plt.ylabel("Tempeture")
    plt.xlabel("Itteration")
    plt.figure(2)
    plt.plot(np.array(zgraph))
    plt.ylabel("z")
    plt.xlabel("Itteration")
    plt.figure(3)
    plt.plot(np.array(ugraph))
    plt.ylabel("u")
    plt.xlabel("Itteration")


def Q2():
    T = 100  #initial temperature
    Tmin = 0.00001 #turminate temperature
    u = 0  
    ubest = 0
    v = 0
    vbest = 0
    z = generatevalue(u,v,1,1)
    zbest = generatevalue(u,v,1,1)
    t = 0  # time spend
    t1 = 0
    dr = 0.99 
    Tgraph = []
    ugraph = []
    vgraph = []
    zgraph = []
    
    while T > Tmin :
        z = generatevalue(u,v,1,1)
        uNew = u + np.random.uniform(low = -0.2, high = 0.2)
        vNew = v + np.random.uniform(low = -0.2, high = 0.2)
        if ((uNew<=1) & (uNew>=-1)&(vNew<=1) & (vNew>=-1)):    
            zNew = generatevalue(uNew,vNew,1,1) #create new z value
            if zNew > z:
                u = uNew #always accept
                v = vNew
            else:
                p = math.exp((zNew -zbest)/T)
                r = np.random.uniform(low = 0, high = 1)
                #print ("p: ", p, "   r: ",r)
                if r < p:
                    u = uNew  #accept based on a ratio
                    v = vNew
        
        T = T*dr    # cooling   
        
        #print ("T: ", T, "  u: ", u, "  v: ", v,"  z: ", z)
        if zNew > zbest:
            zbest = zNew
            ubest = uNew
            vbest = vNew
            t1 = t
            
        Tgraph.append(T)
        ugraph.append(u)
        vgraph.append(v)
        zgraph.append(z)
        t += 1
        
    print("Question 2b:")    
    print("Itteration: ", t, "  Tempeture:",T,"  u best:", ubest,"  v best: ",vbest,"  z best: ", zbest, "   get on iteration: ", t1)   
    plt.figure(1)
    plt.plot(np.array(Tgraph))
    plt.ylabel("Tempeture")
    plt.xlabel("Itteration")
    plt.figure(2)
    plt.plot(np.array(zgraph))
    plt.ylabel("z")
    plt.xlabel("Itteration")
    plt.figure(3)
    plt.plot(np.array(ugraph))
    plt.ylabel("u")
    plt.xlabel("Itteration")
    plt.figure(4)
    plt.plot(np.array(vgraph))
    plt.ylabel("v")
    plt.xlabel("Itteration")
    
    
def Q3():
    T = 100  #initial temperature
    Tmin = 0.001 #turminate temperature
    u = 0  
    ubest = 0
    v = 0
    vbest = 0
    w = 0
    wbest = 0
    z = generatevalue(u,v,w,1)
    zbest = generatevalue(u,v,w,1)
    t = 0  # time spend
    t1 = 0
    dr = 0.99
    Tgraph = []
    ugraph = []
    vgraph = []
    wgraph = []
    zgraph = []
    
    while T > Tmin :
        z = generatevalue(u,v,w,1)
        uNew = u + np.random.uniform(low = -0.3, high = 0.3)
        vNew = v + np.random.uniform(low = -0.3, high = 0.3)
        wNew = w + np.random.uniform(low = -0.3, high = 0.3)
        if ((uNew<=1) & (uNew>=-1)&(vNew<=1) & (vNew>=-1)&(wNew<=1) & (wNew>=-1)):    
            zNew = generatevalue(uNew,vNew,wNew,1) #create new z value
            if zNew > z:
                u = uNew #always accept
                v = vNew
                w = wNew
            else:
                p = math.exp((zNew -z)/T)
                r = np.random.uniform(low = 0, high = 1)
                #print ("p: ", p, "   r: ",r)
                if r < p:
                    u = uNew  #accept based on a ratio
                    v = vNew
                    w = wNew
        T = T*dr
    
        #print ("T: ", T, "  u: ", u, "  v: ", v,"  w: ", w,"  z: ", z)
        if zNew > zbest:
            zbest = zNew
            ubest = uNew
            vbest = vNew
            wbest = wNew
            t1 = t
            
        Tgraph.append(T)
        ugraph.append(u)
        vgraph.append(v)
        wgraph.append(w)
        zgraph.append(z)
        t += 1
        
    print("Question 2c:")    
    print("Cooling times: ", t, "  Tempeture:",T,"  u best:", ubest,"  v best: ",vbest,"  w best",wbest,"  z best: ", zbest,"   get on iteration: ", t1)   
    plt.figure(1)
    plt.plot(np.array(Tgraph))
    plt.ylabel("Tempeture")
    plt.xlabel("Itteration")
    plt.figure(2)
    plt.plot(np.array(zgraph))
    plt.ylabel("z")
    plt.xlabel("Itteration")
    plt.figure(3)
    plt.plot(np.array(ugraph))
    plt.ylabel("u")
    plt.xlabel("Itteration")
    plt.figure(4)
    plt.plot(np.array(vgraph))
    plt.ylabel("v")
    plt.xlabel("Itteration")
    plt.figure(5)
    plt.plot(np.array(wgraph))
    plt.ylabel("w")
    plt.xlabel("Itteration")
    


def Q4():
    T = 100  #initial temperature
    Tmin = 0.001 #turminate temperature
    u = 0  
    ubest = 0
    v = 0
    vbest = 0
    w = 0
    wbest = 0
    z = generatevalue(u,v,w,1)
    zbest = generatevalue(u,v,w,1)
    y = 1
    ybest = 1
    t = 0  # time spend
    t1 = 0
    dr = 0.99
    Tgraph = []
    ugraph = []
    vgraph = []
    wgraph = []
    zgraph = []
    
    while T > Tmin :
        z = generatevalue(u,v,w,y)
        uNew = u + np.random.uniform(low = -0.3, high = 0.3)
        vNew = v + np.random.uniform(low = -0.3, high = 0.3)
        wNew = w + np.random.uniform(low = -0.3, high = 0.3)
        if ((uNew<=1) & (uNew>=-1)&(vNew<=1) & (vNew>=-1)&(wNew<=1) & (wNew>=-1)):    
            zNew = generatevalue(uNew,vNew,wNew,1) #create new z value
            if zNew < 0:
                y = -1
                zNew = -zNew
            if zNew > z:
                u = uNew #always accept
                v = vNew
                w = wNew
            else:
                p = math.exp((zNew -z)/T)
                r = np.random.uniform(low = 0, high = 1)
                #print ("p: ", p, "   r: ",r)
                if r < p:
                    u = uNew  #accept based on a ratio
                    v = vNew
                    w = wNew
        T = T*dr
    
        #print ("T: ", T, "  u: ", u, "  v: ", v,"  w: ", w,"  y: ", y,"  z: ", z)
        if zNew > zbest:
            zbest = zNew
            ubest = uNew
            vbest = vNew
            wbest = wNew
            ybest = y
            t1 = t
            
        Tgraph.append(T)
        ugraph.append(u)
        vgraph.append(v)
        wgraph.append(w)
        zgraph.append(z)
        t += 1
        
    print("Question 2d:")    
    print("Cooling times: ", t, "  Tempeture:",T,"  u best:", ubest,"  v best: ",v,"  w best",w,"  y best",ybest, "  z best: ", zbest,"   get on iteration: ", t1)   
    plt.figure(1)
    plt.plot(np.array(Tgraph))
    plt.ylabel("Tempeture")
    plt.xlabel("Itteration")
    plt.figure(2)
    plt.plot(np.array(zgraph))
    plt.ylabel("z")
    plt.xlabel("Itteration")
    plt.figure(3)
    plt.plot(np.array(ugraph))
    plt.ylabel("u")
    plt.xlabel("Itteration")
    plt.figure(4)
    plt.plot(np.array(vgraph))
    plt.ylabel("v")
    plt.xlabel("Itteration")
    plt.figure(5)
    plt.plot(np.array(wgraph))
    plt.ylabel("w")
    plt.xlabel("Itteration")
    



```
